```python
# Calculate the correlation matrix
corr_matrix = X.corr()

# Unstack the correlation matrix
corr_unstacked = corr_matrix.unstack()
```

```python
# Remove self-correlations (correlations of a feature with itself)
corr_unstacked = corr_unstacked[corr_unstacked.index.get_level_values(0) != corr_unstacked.index.get_level_values(1)]

# Sort by the absolute value of correlations in descending order
top_correlations = corr_unstacked.abs().sort_values(ascending=False).head(7)

# Get the indices of the top correlations
top_indices = top_correlations.index

# Create a new correlation matrix with only the top correlations
filtered_corr_matrix = pd.DataFrame(index=set(top_indices.get_level_values(0)), columns=set(top_indices.get_level_values(1)))
for i, j in top_indices:
    filtered_corr_matrix.loc[i, j] = corr_matrix.loc[i, j]

# Fill NaN values with 0 to avoid errors in heatmap plotting
filtered_corr_matrix = filtered_corr_matrix.fillna(0)

# Create the heatmap
sns.heatmap(filtered_corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')

# Display the heatmap
plt.show()
```
